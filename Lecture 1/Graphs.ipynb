```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Undirected Graph

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(2, 4)
G.add_edge(3, 4)

# explicitly set positions for plotting
pos = {1: (0, 0), 2: (1, 0), 3: (0.25, -1), 4: (1,-1.25)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 3,
    "width": 5,
}
nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
# show adjacency matrix
nx.to_numpy_array(G)
```

```python
# the adjacency matrix is given as a sparse array
nx.adjacency_matrix(G)
```

# Directed Graph

```python
D = nx.DiGraph()
D.add_nodes_from([1,2,3, 4])
D.add_edges_from([(2,1), (2,3), (2,4), (4,3)])

# explicitly set positions for plotting
pos = {1: (0, 0), 2: (1, 0), 3: (0.25, -1), 4: (1,-1.25)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 3,
    "width": 5,
}
nx.draw_networkx(D, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
# show adjacency matrix
nx.to_numpy_array(D)
```
